```python
def bubbleSort(arr):
    n = len(arr)
 
    # Traverse through all array elements
    for i in range(n):
 
        # Last i elements are already in place
        for j in range(0, n-i-1):
#             print(i,"||",arr[j], "|", arr[j+1], end="")
            datax = arr.copy()
            datax [j]="{"+str(datax [j])
            datax [j+1]=str(datax [j+1])+"}"
            
            if arr[j] > arr[j+1] :
                print("")
                datax = arr.copy()
                datax [j]="{"+str(datax [j])
                datax [j+1]=str(datax [j+1])+"}"
                
                print("                           ", datax)
                print(i,"#",arr[j], "<=", arr[j+1], end="")
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
                data = arr.copy()
                data [j]="{"+str(data [j])
                data [j+1]=str(data [j+1])+"}"
                
                print(" = False    >>>>", data)
                print("")
            else:
                print(i,"#",arr[j], "<=", arr[j+1], " = True", "        ", datax)
#                 print("")
    return arr
```
